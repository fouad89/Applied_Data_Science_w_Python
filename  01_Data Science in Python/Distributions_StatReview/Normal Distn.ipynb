```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from scipy import stats
%matplotlib inline
```

### Using Scipy

```python
# creating a normal distribution with mean=0, std=1
mean =0
std = 1
x = np.arange(-4,4,0.01)
y = stats.norm.pdf(x,mean,std)

plt.plot(x,y)
```

### Using Numpy

```python
mu = 0
sigma = 0.1

#normal distn with sample size 1000
norm_set = np.random.normal(mu,sigma,1000)
plt.hist(norm_set,bins=50);
```
