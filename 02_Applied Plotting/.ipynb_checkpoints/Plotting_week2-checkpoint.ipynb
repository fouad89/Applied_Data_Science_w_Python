```python
#zip generator and unpacking
generator = zip(np.arange(5),np.arange(5,10))
generator
```

```python
list(generator)
```

```python
#unpack
x,y =zip(*generator)
```

```python
print(x,y)
```

```python
#splitting the data and plotting with labels
plt.figure()
plt.scatter(x[:2],y[:2], s =100, c = 'blue',label = 'First Part')
plt.scatter(x[2:],y[2:], s =100, c = 'red', label='Second Part')

#loc will place the legend
plt.legend(loc=4, frameon=False,title='Legend')
```

```python
#axis
plt.xlabel('First part')
plt.ylabel('Second')
plt.title('Data Scatters')
```

```python
from matplotlib.artist import Artist

def rec_gc(art,depth=0):
    if isinstance(art,Artist):
        print(' '*depth +str(art))
        for child in art.get_children():
            rec_gc(child,depth+2)

        
        
        
rec_gc(legend)
```

### Line Plots

```python
linear_data = np.arange(8)
quadratic = linear_data**2

plt.figure()

plt.plot(linear_data,'-o',quadratic,'-o')

plt.plot([10,25,30],'--r')
```

```python
#adding labels and legend
plt.xlabel('Data1')
plt.ylabel('Other')
plt.legend(['Linear','Quad','Random'])
```

```python
#fill between 
#the first three arguments if passed, will highlight only area between the 
#two curves
#if only two arguments passed, then it will highlight all area under the 
#curve to the x-axis
plt.gca().fill_between(linear_data,linear_data,quadratic,
                       facecolor='purple',alpha=0.25)
```

```python
observations_dates = np.arange('2017-01-01','2017-01-09',dtype='datetime64[D]')
observations_dates
```

```python
#converting to pandas datetime format

observations_dates = list(map(pd.to_datetime,observations_dates))
observations_dates
```

```python
plt.figure()
plt.plot(observations_dates,linear_data,'-o',
        observations_dates,quadratic,'-r')
```

```python
#to clean up the the dates axis from overlapping
ax = plt.gca().get_children()[6]
ax
```

```python
plt.figure()
plt.plot(observations_dates,linear_data,'-o',
        observations_dates,quadratic,'-r')
```

```python
#rotating the xaxis 
x_axis = plt.gca().xaxis

for item in x_axis.get_ticklabels():
    item.set_rotation(45)
```

```python
#to adjust the xlabels to be fully nice 
plt.subplots_adjust(bottom=0.25)
```

```python
ax = plt.gca()
ax.set_xlabel('Dates')
ax.set_ylabel('Units')
#adding Latex

ax.set_title('Quadratic ($x^2$) vs. Linear ($x$)')
```

### Bar Charts

```python
plt.figure()
plt.bar(range(len(linear_data)),linear_data,width=0.3,label = 'Linear $x$')
```

```python
#adding second bar
xvals=[]
for item in linear_data:
    xvals.append(item+0.3) #so the new bar doesnt overlap the previous one

plt.bar(xvals,quadratic,width=0.3,label = 'Quadratic $x^2$')
```

```python
#adding labels, info practice 
ax = plt.gca()
ax.set_xlabel('Xvalues')
ax.set_ylabel('Yvalues')
```

```python
ax.legend()
```

```python
ax.set_title('Bar Chart')
```

```python
#stacking bars
plt.figure()
plt.bar(linear_data,linear_data,width=0.3,color='b')
plt.bar(linear_data,quadratic,width=0.3,bottom=linear_data,color='r')
```

```python
#switchign axes
plt.figure()
plt.barh(linear_data,linear_data,height=0.3,color='b')
plt.barh(linear_data,quadratic,height=0.3,left=linear_data,color='r')
```

## Languages Problem

```python
languages = ['Python','SQL','Java','C++','JavaScript']
pos = np.arange(len(languages))
popularity = [56,39,34,34,29]
```

```python
plt.figure()
plt.bar(pos,popularity, align='center')
```

```python
plt.xticks(pos,languages,alpha=0.8)
```

```python
plt.ylabel('% Popularity',alpha=0.8)
```

```python
plt.title('Most Popular Languages on StackOverFlow',alpha=0.8)
```

```python
#remove the yticks,ylabels and xticks
plt.tick_params(top='off', bottom='off', left='off', right='off', labelleft='off', labelbottom='on')
```

```python
#removing frame from around the plot
ax = plt.gca()
ax.set_frame_on(False)
```

```python
#another method to remove borders
for spine in plt.gca().spines.values():
    spine.set_visible(False)
```

```python
#setting the other bars grey
other_bars = plt.gcf().get_children()[1].get_children()[1:5]
for bar in other_bars:
    bar.set_color('grey')
```

```python
#changing all text into grey
plt.gcf().get_children()[1].get_children()[0]
```

```python
#changing all text into grey


all_txt = plt.gcf().get_children()[1].get_children()[-4:-1]
```

```python
for text in all_txt:
    text.set_color('grey')
```

```python
#another method to #setting the other bars grey

# change the bar color to be less bright blue
bars = plt.bar(pos, popularity, align='center', linewidth=0, color='lightslategrey')
# make one bar, the python bar, a contrasting color
bars[0].set_color('#1F77B4')
```

```python
#adding the yvalues on the bars 
# direct label each bar with Y axis values
for bar in bars:
    plt.gca().text(bar.get_x() + bar.get_width()/2, bar.get_height() - 5, str(int(bar.get_height())) + '%', 
                 ha='center', color='w', fontsize=11)
```
